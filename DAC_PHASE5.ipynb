```python
replacements = {
   'City/Town/Village/Area': {
      r'Coimbatore': 'Kovai', 
   }
}

data.replace(replacements, regex=True, inplace=True)
```

```python
data[['SO2','City/Town/Village/Area']].groupby(["City/Town/Village/Area"]).median().sort_values(by='SO2',ascending=False).head(10).plot.bar(color='r')
plt.show()
```

```python
data[['NO2','City/Town/Village/Area']].groupby(["City/Town/Village/Area"]).median().sort_values(by='NO2',ascending=False).tail(10).plot.bar(color='b')
plt.show()
```

```python
data[['RSPM/PM10','City/Town/Village/Area']].groupby(["City/Town/Village/Area"]).median().sort_values(by='RSPM/PM10',ascending=False).tail(10).plot.bar(color='b')
plt.show()
```

```python
data[['PM 2.5','City/Town/Village/Area']].groupby(["City/Town/Village/Area"]).median().sort_values(by='PM 2.5',ascending=False).tail(10).plot.bar(color='g')
plt.show()
```

```python
sns.jointplot(x='SO2', y='NO2', data=data,kind='hex',color='k',xlim={0,100}, ylim={0,100})
```
